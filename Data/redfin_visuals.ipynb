```python
 # Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read CSVs
median_prices = pd.read_csv("../Resources/sales_price_clean_csv")
median_prices
```

```python
median_prices.plot(x="Month", y=["National", "Austin, TX","Idaho Falls, ID","Pine Bluff, AR","Shreveport,LA"], xlabel="Year",
         ylabel="Median House price ($1,000s)", title="Median Home Sale Price")

#Set tickmarks to years
x1=[0,11,23,35,47,59]
labels = [2019, 2020, 2021, 2022, 2023, 2024]
plt.xticks(x1, labels,rotation=45)
#Save figure
plt.savefig("med_home_price.png",bbox_inches = "tight")
plt.show()
```

```python
#Read CSVs
sales_qtys = pd.read_csv("../Resources/sales_qty_clean_csv")
sales_qtys
```

```python
import scipy.stats as st
from scipy.stats import linregress

sales_qtys.plot(x="Month", y="National", xlabel="Year",
         ylabel="Number of Homes Sold", title="Homes Sold")

#Set tickmarks to years
x1=[0,11,23,35,47,59]
labels = [2019, 2020, 2021, 2022, 2023, 2024]
plt.xticks(x1, labels,rotation=45)

# Calculate the correlation coefficient and a linear regression model 
# correlation = round(st.pearsonr(sales_qtys["Month"],sales_qtys["National"])[0],2)

x_values = sales_qtys["Month"]
y_values = sales_qtys["National"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")

#Save figure
plt.savefig("national_home_sales_qty.png",bbox_inches = "tight")
plt.show()
```
