```python
header_names=['Quarter','MORTGAGE30US']
Morgate_rate_df = pd.read_csv("../Resources/Mortage_rate_30_Years.csv",header=None,skiprows=1, names=header_names)
Morgate_rate_df 
```

```python
investors_sales_df = pd.read_csv("../Resources/investor_purchases.csv", encoding="utf8", na_values=[" ","  "])
investors_sales_df.head()
```

```python
combined_data_df = pd.merge(Morgate_rate_df,investors_sales_df, on="Quarter")
combined_data_df.head(100)
```

```python
combined_data_df.to_csv('../combined_data_df.csv')
```

```python
Quarter = pd.read_csv('../combined_data_df.csv')['Quarter']
MORTGAGE30US = pd.read_csv('../combined_data_df.csv')['MORTGAGE30US']
Investor_Purchases = pd.read_csv('../combined_data_df.csv')['Investor Purchases']
All_Home_Sales = pd.read_csv('../combined_data_df.csv')['All Home Sales']
Investor_Market_Share = pd.read_csv('../combined_data_df.csv')['Investor Market Share']
```

```python
COLOR_MORTGAGE30US = "#69b3a2"
COLOR_Investor_Purchases = "#3399e6"

fig, ax1 = plt.subplots(figsize=(9, 9))
ax2 = ax1.twinx()

ax1.plot(Quarter, MORTGAGE30US, color=COLOR_MORTGAGE30US, lw=3)
ax2.plot(Quarter, Investor_Purchases, color=COLOR_Investor_Purchases, lw=4)

ax1.set_xlabel("Date")
ax1.set_ylabel("Mortgage Rate", color=COLOR_MORTGAGE30US, fontsize=14)
ax1.tick_params(axis="y", labelcolor=COLOR_MORTGAGE30US)

ax2.set_ylabel("Investor Purchases", color=COLOR_Investor_Purchases, fontsize=14)
ax2.tick_params(axis="y", labelcolor=COLOR_Investor_Purchases)

fig.suptitle("Mortgage Rate vs Investor Purchases", fontsize=20)
fig.autofmt_xdate()


plt.savefig('Mortgage Rate vs Investor Purchases.png')
plt.show()
```
