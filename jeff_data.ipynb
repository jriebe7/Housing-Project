```python
import requests
import json
import prettytable
import pandas as pd
import numpy as np
import hvplot.pandas
import matplotlib.pyplot as plt

```

```python

mountain_cpi = pd.read_csv("Project 1 Data Files/Mountain CPI.csv", na_values=[" ","  "])
west_sc_cpi = pd.read_csv("Project 1 Data Files/West SC CPI.csv", na_values=[" ","  "])
sales_prices = pd.read_csv("Housing-Project/sales_price_clean_csv")

#mountain_cpi = mountain_cpi.set_index("Year")
# mountain_cpi = np.transpose(mountain_cpi)


#mountain_cpi.plot(kind="line", )


```

```python
df = mountain_cpi.set_index("Year").T
df2 =west_sc_cpi.set_index("Year").T
df
```

```python
df.plot()
plt.ylabel("CPI")
plt.show()
```

```python
df2.plot()
plt.ylabel("CPI")
plt.show()
```

```python
df3 = pd.concat([df[2019],df[2020]])
df4 = pd.concat([df3,df[2021]])
df5 = pd.concat([df4,df[2022]])
df6 = pd.concat([df5, df[2023]])
df7 = pd.concat([df6,df[2024]])

df7 = pd.DataFrame(df7)
df7
```

```python
cpi = []

for x in range(0, len(df7)):
    price = df7.iloc[x,0]
    cpi.append(price)

cpi
```

```python
mountain_cpi_all = pd.DataFrame(cpi)

mountain_cpi_all.columns = ["Mountain Region CPI"]

plt.plot(mountain_cpi_all)

```

```python
df8 = pd.concat([df2[2019],df2[2020]])
df9 = pd.concat([df8,df2[2021]])
df10 = pd.concat([df9,df2[2022]])
df11 = pd.concat([df10, df2[2023]])
df12 = pd.concat([df11,df2[2024]])

df12 = pd.DataFrame(df12)

```

```python
cpi2 = []

for x in range(0, len(df12)):
    price2 = df12.iloc[x,0]
    cpi2.append(price2)

cpi2

west_sc_cpi_all = pd.DataFrame(cpi2)

west_sc_cpi_all.columns = ["West South Central CPI"]
west_sc_cpi_all.plot()
x1 = [0,11,23,35,47,59]
labels = [2019,2020,2021,2022,2023,2024]
plt.xticks(x1, labels, rotation=45)
plt.ylabel("CPI")

plt.savefig("West South Central CPI Graph.png")
plt.show()
```
