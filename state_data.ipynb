Cleaning Median Sales price Data

```python
#Import dependencies
import pandas as pd
import os
import csv
from datetime import datetime
```

```python
with open("Resources/med_sale_price.csv") as f:
    print(f)
```

```python
#Median Sale Price csv into DataFrame
sales_price_df = pd.read_csv("Resources/med_sale_price.csv", encoding='UTF-16', sep="\t").T

#Set row one as headers
new_header =sales_price_df.iloc[0]
sales_price_df=sales_price_df[1:]
sales_price_df.columns =new_header

#Rename some column names
sales_price_df.rename(columns={"Region":"Month", " National":"National"},inplace=True)
sales_price_df
```

```python
sales_price_df["Month"] = pd.to_datetime(sales_price_df["Month"])
```

```python
#Format price to remove text and cast as integers
sales_price_df["National"] = sales_price_df["National"].str.replace("$","")
sales_price_df["National"] = sales_price_df["National"].str.replace("K","")
sales_price_df["National int"] = sales_price_df["National"].astype(int)
sales_price_df["Austin, TX"] = sales_price_df["Austin, TX"].str.replace("$","")
sales_price_df["Austin, TX"] = sales_price_df["Austin, TX"].str.replace("K","")
sales_price_df["Austin int"] = sales_price_df["Austin, TX"].astype(int)
sales_price_df["Idaho Falls, ID"] = sales_price_df["Idaho Falls, ID"].str.replace("$","")
sales_price_df["Idaho Falls, ID"] = sales_price_df["Idaho Falls, ID"].str.replace("K","")
sales_price_df["Idaho Falls int"] = sales_price_df["Idaho Falls, ID"].astype(int)
sales_price_df["Pine Bluff, AR"] = sales_price_df["Pine Bluff, AR"].str.replace("$","")
sales_price_df["Pine Bluff, AR"] = sales_price_df["Pine Bluff, AR"].str.replace("K","")
sales_price_df["Pine Bluff int"] = sales_price_df["Pine Bluff, AR"].astype(int)
sales_price_df["Shreveport, LA"] = sales_price_df["Shreveport, LA"].str.replace("$","")
sales_price_df["Shreveport, LA"] = sales_price_df["Shreveport, LA"].str.replace("K","")
sales_price_df["Shreveport int"] = sales_price_df["Shreveport, LA"].astype(int)
```

```python
sales_price_df.dtypes
```

```python
#Create new clean dataframe
#Remove columns as object types
sales_price_clean = sales_price_df.drop(columns=['National', 'Austin, TX','Idaho Falls, ID','Pine Bluff, AR','Shreveport, LA'])

#Rename columns
sales_price_clean.rename(columns={"National int": "National", "Austin int":"Austin, TX","Idaho Falls int":"Idaho Falls, ID","Pine Bluff int":"Pine Bluff, AR","Shreveport int":"Shreveport,LA"},inplace=True)
sales_price_clean
```

```python
#Write new CSV to repo
sales_price_clean.to_csv("sales_price_clean_csv", index=False, header=True)
```

```python
#Home Sales csv into DataFrame
sales_qty = pd.read_csv("Resources/sales (1).csv", encoding='UTF-16', sep="\t")

#Rename some column names

sales_qty=sales_qty.T

#Set row one as headers
new_header =sales_qty.iloc[0]
sales_qty=sales_qty[1:]
sales_qty.columns =new_header
sales_qty_index = sales_qty.reset_index()
sales_qty_index.rename(columns={"index":"Month", " National":"National"},inplace=True)
sales_qty_index
```

```python
#Cast month as datetime
sales_qty_index["Month"] = pd.to_datetime(sales_qty_index["Month"])
sales_qty_index
```

```python
#Format qty to remove some text and to cast as integers
sales_qty_index["National"] = sales_qty_index["National"].str.replace(",","")
sales_qty_index["National int"] = sales_qty_index["National"].astype(int)
sales_qty_index["Austin, TX"] = sales_qty_index["Austin, TX"].str.replace(",","")
sales_qty_index["Austin int"] = sales_qty_index["Austin, TX"].astype(int)
sales_qty_index["Idaho Falls, ID"] = sales_qty_index["Idaho Falls, ID"].str.replace(",","")
sales_qty_index["Idaho Falls int"] = sales_qty_index["Idaho Falls, ID"].astype(int)
sales_qty_index["Pine Bluff, AR"] = sales_qty_index["Pine Bluff, AR"].str.replace(",","")
sales_qty_index["Pine Bluff int"] = sales_qty_index["Pine Bluff, AR"].astype(int)
sales_qty_index["Shreveport, LA"] = sales_qty_index["Shreveport, LA"].str.replace(",","")
sales_qty_index["Shreveport int"] = sales_qty_index["Shreveport, LA"].astype(int)
```

```python
#Create new clean dataframe
#Remove columns as object types
sales_qty_clean = sales_qty_index.drop(columns=['National', 'Austin, TX','Idaho Falls, ID','Pine Bluff, AR','Shreveport, LA'])

#Rename columns
sales_qty_clean.rename(columns={"National int": "National", "Austin int":"Austin, TX","Idaho Falls int":"Idaho Falls, ID","Pine Bluff int":"Pine Bluff, AR","Shreveport int":"Shreveport,LA"},inplace=True)
sales_qty_clean
```

```python
#Write new CSV to repo
sales_qty_clean.to_csv("sales_qty_clean_csv", index=False, header=True)
```
